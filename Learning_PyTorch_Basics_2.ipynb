```python
import torch
```

## **Tensor Shapes**

```python
x = torch.tensor([[1,2,3],[4,5,6]])
x
```

```python
x.shape
```

**if we need another tensor of shape similar to any tensor**

```python
torch.empty_like(x)
```

```python
torch.ones_like(x)
#similar with zeros_like
```

```python
# torch.rand_like(x) this doesn't work becacuse of mismatching of data types
torch.rand_like(x, dtype=torch.float32)
```

## **Tensor DataTypes**

```python
#find data type
x.dtype
```

```python
#assign data type
torch.tensor([1.0,2.0,3.0], dtype=torch.int32)
```

```python
torch.tensor([1,2,3], dtype=torch.float64)
```

```python
#using "to"
x.to(torch.float32) #changes dtype of tensor
```

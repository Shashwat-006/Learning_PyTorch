```python
import torch
from torch import nn
import pandas as pd
```

```python
dataset = pd.read_csv('/content/diabetes.csv')
```

```python
dataset.head()
```

```python
dataset['Outcome'].value_counts()
```

```python
y = dataset['Outcome']
x = dataset.drop(columns = ['Outcome'])
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,  stratify = y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
x_train = scalar.fit_transform(x_train)
```

```python
x_test = scalar.transform(x_test)
```

```python
print(type(x_train))
```

```python
x_train = torch.tensor(x_train , dtype = torch.float32)
x_test = torch.tensor(x_test , dtype = torch.float32)
y_train = torch.tensor(y_train.values, dtype = torch.float32)
y_test = torch.tensor(y_test.values, dtype = torch.float32)
```

```python
model = nn.Sequential(
    nn.Linear(8, 1),
    nn.Sigmoid()
)
```

```python
loss_fn = nn.BCELoss()
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
```

```python
print(model[0].weight)
```

```python
for epoch in range (200):
  y_pred = model(x_train).squeeze()
  loss = loss_fn(y_pred, y_train)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch % 10 == 0:
    print(f'Epoch {epoch}, Loss {loss.item():.4f}')
```

```python
print(model[0].weight)
```

```python
model.eval()

with torch.no_grad():
  train_pred = model(x_train)
  train_pred = (train_pred > 0.5).float()
  test_pred = model(x_test)
  test_pred = (test_pred > 0.5).float()


  train_acc = (train_pred.squeeze() == y_train).float().mean()
  print(f'Train Accuracy {train_acc.item()*100:.2f}')

  test_acc = (test_pred.squeeze() == y_test).float().mean()
  print(f'Test Accuracy {test_acc.item()*100:.2f}')
```

This is just for learning how the pipeline for a neural network works in PyTorch from basics by coding everything manually rather than using modules provided by Pytorch

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

# **Basic Operations for the Model**

```python
data = pd.read_csv('/content/breast-cancer.csv')
```

```python
data.head()
```

```python
Y = data['diagnosis']
X = data.drop(columns =['id','diagnosis'],axis=1)
```

```python
X.head()
```

```python
X.shape
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

Standardizing the data

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

Converting diagnosis column to numerical values

```python
encoder = LabelEncoder()
```

```python
Y_train = encoder.fit_transform(Y_train)
Y_test = encoder.transform(Y_test)
```

```python
Y_train
```

Numpy arrays to PyTorch TENSORS

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
Y_train_tensor = torch.from_numpy(Y_train)
Y_test_tensor = torch.from_numpy(Y_test)
```

```python
X_train_tensor
```

# **Defining the Model**

```python
class firstNN():
  def __init__(self,X):
    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True) #taking a random value at first then will change it accordingly
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y): #binary cross entropy loss
    epsilon = 1e-7 #avoiding log(0) condition
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    #loss calculation
    loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss
```

# **Important Parameters**

```python
learning_rate = 0.1
no_of_epochs = 40
```

# **Training Pipeline**

```python
#creating model
model = firstNN(X_train_tensor)

for epochs in range(no_of_epochs):

  #forward pass
  y_pred = model.forward(X_train_tensor)

  #loss calculation
  loss = model.loss_function(y_pred, Y_train_tensor)

  #backward pass
  loss.backward()

  #parameters updation
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  #zero gradients so that gradients doesn't accumulate
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  #visualizing loss with each epoch
  print(f'Epoch: {epochs+1}, Loss: {loss.item()}')
```

```python
model.bias
```

# **Evaluation**

```python
#model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor).round()
  accuracy = (y_pred == Y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy}')
```

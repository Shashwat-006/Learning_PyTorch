# **NN Module**

```python
import torch
import torch.nn as nn
```

## **Example 1.** assuming a binary classification problem that has 5 features and gives 1 output & uses sigmoid activation function.

```python
#creating model class


class Model(nn.Module):
  def __init__(self, num_of_features):

    super().__init__()
    self.linear = nn.Linear(num_of_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)

    return out
```

```python
#creating a dummy dataset
dataset = torch.rand(10,5)

#create model
model = Model(dataset.shape[1])

#calling model for forward pass
model(dataset)    #model.forward(dataset) this can also work but this isn't standard way of calling forward pass in PyTorch
```

# **Working on Breast Cancer Data using NN Module**

Working on the same which I hard coded manually without using any modules previously

## **Pre-Processing**

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('/content/breast-cancer.csv')
```

```python
Y = data['diagnosis']
X = data.drop(columns =['id','diagnosis'],axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
encoder = LabelEncoder()
```

```python
Y_train = encoder.fit_transform(Y_train)
Y_test = encoder.transform(Y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_train_tensor = X_train_tensor.float()
X_test_tensor = torch.from_numpy(X_test)
X_test_tensor = X_test_tensor.float()
Y_train_tensor = torch.from_numpy(Y_train)
Y_train_tensor = Y_train_tensor.float()
Y_test_tensor = torch.from_numpy(Y_test)
Y_test_tensor = Y_test_tensor.float()
```

## **UPDATION w.r.t previous**

```python
class firstNN(nn.Module):
  def __init__(self, num_of_features):
    super().__init__()
    self.linear = nn.Linear(num_of_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out


```

```python
learning_rate = 0.1
no_of_epochs = 40
```

```python
# removing manual loss fn from class definition and using module's loss fn
loss_function = nn.BCELoss()
```

```python
#creating model
model = firstNN(X_train_tensor.shape[1])

# using module's optimizer and removing manual one
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epochs in range(no_of_epochs):

  #forward pass
  y_pred = model(X_train_tensor)

  #loss calculation
  loss = loss_function(y_pred, Y_train_tensor.view(-1,1))

  #zero gradients so that gradients doesn't accumulate
  optimizer.zero_grad()

  #backward pass
  loss.backward()

  #parameters updation
  optimizer.step()

  #visualizing loss with each epoch
  print(f'Epoch: {epochs+1}, Loss: {loss.item()}')
```

```python
model.linear.weight
```

```python
#model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor).round()
  accuracy = (y_pred == Y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy}')
```

```python
#using zeros
torch.zeros(2,4)

#similar with ones
```

```python
#using rand
torch.rand(2,3)
```

```python
#use of seed
torch.rand(2,4)
```

```python
#manual seed
torch.manual_seed(42)
torch.rand(2,4)
```

On using manual seed, we'll get same random values

```python
torch.manual_seed(42)
torch.rand(2,4)
```

```python
#using tensor
torch.tensor([[1,2,3],[4,5,6]])
```

```python
#other ways

#arange
print(torch.arange(1,11))

#using linspace
print(torch.linspace(0,10,10)) #0 se 10 tk ten equi-distant values

#using eye
print(torch.eye(3))

#using full
print(torch.full((2,3),6))
```

# **Copying a TENSOR**

```python
a = torch.rand(2,3)
a
```

```python
b = a #there's a problem with this
b
```

```python
a[0][0] = 0 #if we do some change in a, it's occurs in b as well
b
```

```python
#this is how we prove it
print(id(a))
print(id(b))
```

This means that assignment operator ( = ), doesn't copy rather it makes the variable to refer to the location where the tensor is stored

**This is how we copy**

```python
b = a.clone()
b
```

```python
a[0][1] = 0
print(a)
print(b)
```

# **Tensor Operations on GPU**

```python
torch.cuda.is_available()
```

```python
device = torch.device('cuda')
```

```python
#creating new tensor on GPU
torch.rand((2,3), device=device)
```

```python
a = torch.rand(2,3)
a
```

```python
#moving an existing tensor to GPU
a.to(device)
```

# **Reshaping Tensors**

```python
a = torch.ones(4,4)
a
```

```python
#reshape
a.reshape(2,2,4) #or (2,2,2,2)
```

```python
#flatten
a.flatten()
```

```python
b = torch.rand(2,3,4)
b
```

```python
#permute
b.permute(2,0,1) #there 2,0,1 are indices of shape of the tensor
```

```python
#unsqeeze
#typical image size = (216,216,3)

c=torch.rand(216,216,3)
c.unsqueeze(0).shape   #We pass the index where new dim (1) is to be added as argument
```

```python
#squeeze -> this only removes a dim if dimension at given argument is 1

d = torch.rand(2,20)
print(d.squeeze(0).shape)

e = torch.rand(3,1,3)
print(e.squeeze(1).shape)
```

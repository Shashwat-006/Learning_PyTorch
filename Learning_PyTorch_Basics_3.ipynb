# **Mathematical Operations**

## 1. Scalar Operations

```python
x =torch.rand(2,2)
x
```

```python
#addition
print(x + 2)

#subtraction
print(x - 2)

#multiplication
print(x * 2)

#division
print(x / 2)

#int division
print((x*100)//3)

#mod
print(((x*100)//3)%5)

#power
print(x**3)
```

## 2. Element-wise Operation

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
# add
# torch.add(a,b) or
a + b

# subtract
a - b

# multiply
a * b

# division
a / b

# mod
a % b

# power
a ** b
```

```python
c = torch.tensor([1, -2, 3, -4])

# abs (absolute value)
torch.abs(c)
```

```python
#negative
torch.neg(c)
```

```python
d = torch.tensor([1.9, 2.3, 3.7, 4.4])

#round
torch.round(d)
```

```python
#ceil
torch.ceil(d)
```

```python
#floor
torch.floor(d)
```

```python
#clamp
torch.clamp(d, min=2, max=3) #elements less than 'min' becomes 'min' & greater than 'max' becomes 'max'
```

## 3. Reduction Operation

```python
e = torch.randint(size=(2,3), low=0, high=10)
e
```

```python
#sum
torch.sum(e)

#sum along column
torch.sum(e, dim=0)

#sum along row
torch.sum(e, dim=1)
```

```python
f = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)

#mean
torch.mean(f) #only supports floating or complex no.

#mean along col
torch.mean(f, dim=0)
```

```python
#median
torch.median(e)
```

```python
#max and min
torch.max(e)
torch.min(e, dim=1)
```

```python
#product
torch.prod(e)
```

```python
#std deviation
torch.std(f) #only supports floating or complex no.
```

```python
#variance
torch.var(f, dim=0)
```

```python
#argmax
torch.argmax(e) #gives index of highest no.
```

```python
#argmin
torch.argmin(e)
```

## 4. Matrix Operations

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
```

```python
#matrix multiplication

torch.matmul(f,g)

#torch.matmul(g,f)
```

```python
vector1 = torch.tensor([1,2])
vector2 = torch.tensor([3,4])

#dot product
torch.dot(vector1, vector2)
```

```python
#transpose
torch.transpose(f, 0, 1) #(matrix , dim1, dim2 (kis dim ko kis dim me krna h))
```

```python
h = torch.randint(0, 10, (3,3)).to(torch.float32) #randint accepts arg as follows (low, high, size) by default
h
```

```python
#determinant
torch.det(h)
```

```python
#inverse
torch.inverse(h)
```

## 5. Comparision Opertaions

```python
i = torch.randint(0, 10, (2,3))
j = torch.randint(0, 10, (2,3))

print(i)
print(j)
```

```python
#greater than
i > j

#less than
i < j

#equal to
i == j

#Not equal to
i != j

#less than equal to
i <= j

#greater than equal to
i >= j
```

# **Inplace Operations**

```python
m = torch.rand(2,3)
n = torch.rand(2,3)

print(m)
print(n)
```

underscore ( _ ) represents inplace operation

```python
m.add_(n) #add m and n & stores the result in 'm'
print(m)
```

```python
m.relu_() #stores the result after applying relu in 'm'
print(m)
```
